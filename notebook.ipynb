```python
clean_values = []
for price in values['Value']:
    
    clean_price = str(price).replace("€", "")
    
    if 'm' in clean_price:
        number = float(clean_price.replace('m', '')) * 1_000_00
    elif 'k' in clean_price:
        number = float(clean_price.replace('k', '')) * 1_000
    else:
        number = 0
    clean_values.append(number)

values['Value_Euros'] = clean_values
print(values["Value_Euros"])
```

```python
cols = list(stats.columns)
cols[2] = 'League'
cols[3] = 'Season'
cols[4] = 'team'    
cols[5] = 'player'  
cols[6] = 'Nation'
cols[7] = 'Pos'
cols[8] = 'Age'
cols[9] = 'Born'
stats.columns = cols
stats = stats.drop(columns=['level_0', 'index'], errors='ignore')
print(stats.head())
```

```python
team_mapping = {
    'Man Utd': 'Manchester Utd',
    'Newcastle': 'Newcastle Utd',
    "Nott'm Forest": "Nott'ham Forest",
    'Man City': 'Manchester City',
    'Leeds': 'Leeds United'
}

values['Team'] = values['Team'].replace(team_mapping)

final_df = pd.merge(
    stats,
    values,
    left_on=['player', 'team'],
    right_on=['Name', 'Team'],
    how='inner'
)

final_df.drop(columns=['Name', 'Team'])

print(f"{len(final_df)}")
print(final_df[['player', 'team', 'Value_Euros', 'Gls', 'Ast']].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
```

```python


features = ['age', 'npxg+xag', 'starts', 'team', 'pos']
target = 'value_euros'

model_df = final_df[features + [target]].dropna().copy()

for col in ['team', 'pos']:
    if model_df[col].dtype == 'object':
        model_df[col]  = pd.factorize(model_df[col])[0]   

#Split 60/20/20
df_full_train, df_test = train_test_split(model_df, test_size=0.2, random_state=42)

df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=42)

def prepare_data(df):
    x = df[features].copy()
    y = df[target].values
    return x, y

x_train, y_train = prepare_data(df_train)
x_test, y_test = prepare_data(df_test)
x_val, y_val = prepare_data(df_val)

print(f"{len(x_train)}")
print(f"{len(x_test)}")
print(f"{len(x_val)}")
```

```python
#Train The model
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)

#Predict
y_pred = lr_model.predict(x_val)

#Evaluate
r2 = r2_score(y_val, y_pred)
mae = mean_absolute_error(y_val, y_pred)
rmse = np.sqrt(mean_squared_error(y_val, y_pred))
mape = mean_absolute_percentage_error(y_val, y_pred)

print(f"Accuracy (R2 Score):  {r2:.2%}")
print(f"Average Error (MAE):  €{mae:,.0f} (Lower is better)")
print(f"RMSE (Punishes Big Misses):    €{rmse:,.0f} (Lower is better)")
print(f"MAPE (Avg % Error):            {mape:.1%}     (Lower is better)")
```

```python
from sklearn.tree import DecisionTreeRegressor

#Train the model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(x_train, y_train)

#Predict
y_pred_dt = dt_model.predict(x_val)

#Evaluate
r2_dt = r2_score(y_val, y_pred_dt)
mae_dt = mean_absolute_error(y_val, y_pred_dt)
rmse_dt = np.sqrt(mean_squared_error(y_val, y_pred_dt))
mape_dt = mean_absolute_percentage_error(y_val, y_pred_dt)

print(f"Accuracy (R2 Score):  {r2_dt:.2%}")
print(f"Average Error (MAE):  €{mae_dt:,.0f} (Lower is better)")
print(f"RMSE (Punishes Big Misses):    €{rmse_dt:,.0f} (Lower is better)")
print(f"MAPE (Avg % Error):            {mape_dt:.1%}     (Lower is better)")
```

```python
from sklearn.ensemble import RandomForestRegressor

#Train the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_model.fit(x_train, y_train)

#Predict
y_pred_rf = rf_model.predict(x_val)

#Evaluate
r2_rf = r2_score(y_val, y_pred_rf)
mae_rf = mean_absolute_error(y_val, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_val, y_pred_rf))
mape_rf = mean_absolute_percentage_error(y_val, y_pred_rf)

print(f"Accuracy (R2 Score):  {r2_rf:.2%}")
print(f"Average Error (MAE):  €{mae_rf:,.0f} (Lower is better)")
print(f"RMSE (Punishes Big Misses):    €{rmse_rf:,.0f} (Lower is better)")
print(f"MAPE (Avg % Error):            {mape_rf:.1%}     (Lower is better)")
```

```python
from xgboost import XGBRegressor

#Train the model
xgb_model = XGBRegressor(n_estimators=1000, learning_rate=0.05, early_stopping_rounds=10, random_state=42, n_jobs=-1)
xgb_model.fit(x_train, y_train,
              eval_set=[(x_val, y_val)],
              verbose=False)

#predict
y_pred_xgb = xgb_model.predict(x_val)

#Evaluate
r2_xgb = r2_score(y_val, y_pred_xgb)
mae_xgb = mean_absolute_error(y_val, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_val, y_pred_xgb))
mape_xgb = mean_absolute_percentage_error(y_val, y_pred_xgb)

print(f"Accuracy (R2 Score):  {r2_xgb:.2%}")
print(f"Average Error (MAE):  €{mae_xgb:,.0f} (Lower is better)")
print(f"RMSE (Punishes Big Misses):    €{rmse_xgb:,.0f} (Lower is better)")
print(f"MAPE (Avg % Error):            {mape_xgb:.1%}     (Lower is better)")
```

__Despite Linear Regression having a slightly higher $R^2$, we selected XGBoost as our champion model. Here is the business logic behind this decision:__

__1. Cash is King (MAE vs $R^2$): In a player valuation context, the "Goodness of Fit" ($R^2$) matters less than the actual price tag accuracy.Linear Regression is off by an average of €1.13M per player.XGBoost is off by only €948k per player.Business Impact: Using XGBoost saves the club roughly €180,000 per negotiation in estimation error. Over a transfer window of 10 players, this prevents €1.8M in potential overpayment/undervaluation.__

__2. The "Percentage" Gap (MAPE): Linear Regression struggles significantly with lower-value players, resulting in a 68% MAPE.This means for a €5M player, the Linear model might predict €8.4M, a disaster for a budget club.XGBoost (54.9%) handles these non-linear relationships much better, making it more reliable across the entire squad, not just for the superstars.__

__3. Conclusion: While Linear Regression captures the general trend of the "Superstars" well (high $R^2$), XGBoost is the more robust, financially accurate tool for the majority of the dataset.__

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 500, 1000],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
}

xgb_tuner = XGBRegressor(random_state=42, n_jobs=-1, early_stopping_rounds=10)

grid_search = GridSearchCV(
    estimator=xgb_tuner,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',
    cv=3, 
    verbose=1
)

grid_search.fit(x_train, y_train,
                eval_set=[(x_train, y_train)],
                verbose=False)

best_model = grid_search.best_estimator_
print(f"Best Params Are: {grid_search.best_params_}")

#Validation of tuned model
y_pred_tuned = best_model.predict(x_val)

r2_tuned = r2_score(y_val, y_pred_tuned)
mae_tuned = mean_absolute_error(y_val, y_pred_tuned)
rmse_tuned = np.sqrt(mean_squared_error(y_val, y_pred_tuned))
mape_tuned = mean_absolute_percentage_error(y_val, y_pred_tuned)

print(f"R2 Score: {r2_tuned:.2%} (Default was {r2_xgb:.2%})")
print(f"MAE:      €{mae_tuned:,.0f} (Default was €{mae_xgb:,.0f})") 
print(f"RMSE:     €{rmse_tuned:,.0f} (Default was €{rmse_xgb:,.0f})")
print(f"MAPE:     {mape_tuned:.1%}     (Default was {mape_xgb:.1%})")
```

```python
#Final Prediction
y_pred_final = best_model.predict(x_test)

#Final Evaluation
r2_final = r2_score(y_test, y_pred_final)
mae_final = mean_absolute_error(y_test, y_pred_final)
rmse_final = np.sqrt(mean_squared_error(y_test, y_pred_final))
mape_final = mean_absolute_percentage_error(y_test, y_pred_final)

print(f"R2 Score: {r2_final:.2%}")
print(f"MAE:      €{mae_final:,.0f}")
print(f"RMSE:     €{rmse_final:,.0f}")
print(f"MAPE:     {mape_final:.1%}")
```

__Outlier Detection:__ _The plot highlights specific cases where the model struggled:_
